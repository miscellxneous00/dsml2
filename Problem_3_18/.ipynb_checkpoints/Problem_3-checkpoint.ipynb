```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_path="./House Price Prediction Dataset.csv"
```

```python
df=pd.read_csv(csv_path)
```

```python
df.head()
```

```python
df.info()
```

```python
numeric_cols = df.select_dtypes(exclude=['object']).columns
```

```python
numeric_cols
```

```python
print("=== STANDARD DEVIATION FOR EACH NUMERIC FEATURE ===")
df[numeric_cols].std()
```

```python
print("=== Variance FOR EACH NUMERIC FEATURE ===")
df[numeric_cols].var()
```

```python
print("=== Percentiles FOR EACH NUMERIC FEATURE ===")
df[numeric_cols].quantile([0.25,0.5,0.75])
```

```python
print("Plotting Histograms...")
```

```python
df[numeric_cols].hist(figsize=(15,12),bins=20)
plt.tight_layout()
plt.show()
```

```python
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    plt.hist(df[col].dropna(), bins=20)  # clean and simple
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()
```
